```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Load the dataset
file_path = 'marketing_campaign_dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Clustering campaigns based on performance metrics
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Clicks', 'Impressions', 'Engagement_Score', 'ROI']])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualize clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Clicks'], y=df['ROI'], hue=df['Cluster'], palette='viridis')
plt.title('Campaign Clustering Based on Performance')
plt.xlabel('Clicks')
plt.ylabel('ROI')
plt.show()

# Save cleaned data for further analysis
df.to_csv('clustered_marketing_data.csv', index=False)
```
